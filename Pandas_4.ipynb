# Merge

```python
import pandas as pd
```

```python
left = pd.DataFrame({"id":[1,2,3,4,5],
                    "name":["Vishal", "Rahul", "Shubham", "Jitesh", "Sachin"],
                    "subject_id":["sub1","sub2", "sub3", "sub4", "sub6"]})

right = pd.DataFrame({"id":[1,2,3,4,5],
                     "name":["kartik", "ravi", "satish", "Sanket", "ravi"],
                     "subject_id": ["sub2", "sub3", "sub4", "sub6", "sub5"]})
```

```python
# (left,right) here sequence is matter
# is get none because of which bases we will merge dataframe
# you specified the any column to which bases they merge the data
```

```python
pd.merge(left,right) 
```

```python
# we have to specify on which column you want to perform merge
```

```python
left
```

```python
right
```

```python
# on means specified column in which you want to merge the data
#when we are performing merge operation 
# you must have to give column name on which merge will work
pd.merge(left,right, on=["id"]) 
```

```python
#suffixes _x means left  dataset
#suffixes _y menas right dataset
# We can change the suffixes
pd.merge(left,right, on=["id"],suffixes=("_Left", "_Right"))
```

```python
# lets change the id's of right

right = pd.DataFrame({"id":[1,2,3,4,7],
                     "name":["kartik", "ravi", "satish", "Sanket", "ravindra"],
                     "subject_id": ["sub3", "sub2", "sub3", "sub4", "sub5"]})
left,right
```

```python
# if change the id they merge only similar ids data
# how its possible
# in merge "how" parameter is imporatant
# by defualt how = "inner"
# "inner" means inner join means intersection means common things
# seen below output
```

```python
pd.merge(left, right, on='id')
```

```python
# when we want all the data we should perform 'outer' == means Union means all data
# they merge the all data but they put uncommon entriys NaN values
# how ["inner", "outer", "left", "right"] these are the value we can get how
```

```python
pd.merge(left, right, on='id',how ='outer')
```

```python
# how = "left"  means only left dataframe all keys 
# the id 5 is in left but not in right they put the right id5 == NaN value
pd.merge(left, right, on='id',how ='left') # left join
```

```python
# how = "right"  means only right dataframe all keys 
# id 7 not present in left side datframe then put NaN 
pd.merge(left, right, on='id',how ='right') # right join
```

```python
left,right
```

```python
# Merging with two keys/columns
# common id and subject_id dataframe
pd.merge(left,right, on=["id", "subject_id"])
```

```python
left,right
```

```python
# how = "outer"
pd.merge(left,right, on=["id", "subject_id"], how="outer")
```

```python
# how = "left"
pd.merge(left,right, on=["id", "subject_id"], how="left")
```

```python
right
```

```python
# how = "right"
pd.merge(left,right, on=["id", "subject_id"], how="right")
```

# Join

```python
#Joins on the basis of column
# its join 
# by default How = "left"
# on = "None"
#  lsuffix="_L", rsuffix="_R" its compulsory defined
left.join(right, lsuffix="_L", rsuffix="_R")
```

```python
# on = id
# how = left 
left.join(right,on="id", lsuffix="_L", rsuffix="_R")
```

```python
# how = "right"
left.join(right,on = "id", lsuffix="_L", rsuffix="_R", how="right")
```

```python
# how = "inner" (common)
left.join(right,on = "id", lsuffix="_L", rsuffix="_R", how="inner")
```

```python
# how = "outer"
s= left.join(right,on = "id", lsuffix="_L", rsuffix="_R", how="outer")
```

```python
s.fillna("s",axis=1)
```

# set_index

```python
left.set_index("id").join(right.set_index("id"),on = "id", lsuffix="_L", rsuffix="_R",)
```

# Concatenation
#Concatenation is similar to append

```python
one = pd.DataFrame({ "Name":["Shubham", "vishal", "santosh", "Rahul"],
                   "subject_id":["sub1", "sub2", "sub3", "sub4"],
                   "mark":[42, 55, 65, 75]}, index=[1, 2, 3, 4])
two = pd.DataFrame({"Name":["Satish", "Sam", "monu", "sanket"],
                   "subject_id":["sub2", "sub1", "sub4", "sub3"],
                   "mark":[55, 65, 67, 85]}, index=[1, 2, 3, 4])
one, two
```

```python
pd.concat([one,two],ignore_index=True)
```

```python
#seperate the dataframe using x and y as indicators/keys)
pd.concat([one, two],keys=["1st", "2nd"]) # but ignore index and keys cannot performance in at time
```

```python
pd.concat([one, two],keys=["1st", "2nd"],names=["D_name", "index"])
```

```python
#default axis is 0 means index means rowise
#axis 1 means column means columnise
pd.concat([one, two],keys=["1st", "2nd"],names=["D_name", "index"],axis=1)
```

```python
# 
pd.concat([one,two,one])
```
